# Importing Libraries and Displaying Regions

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the first region
f1 = gpd.read_file('C:\\Users\\LENOVO\\Downloads\\IND_adm\\IND_adm1.shp')

# Load the second region
f2 = gpd.read_file('C:\\Users\\LENOVO\\Downloads\\IND_adm\\IND_adm2.shp')

# Create subplots to plot the maps side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the first region in red with black outline
f1.plot(ax=axes[0], color='red', edgecolor='black')

# Plot the second region in green with black outline
f2.plot(ax=axes[1], color='green', edgecolor='black')

# Set plot titles
axes[0].set_title('Region 1')
axes[1].set_title('Region 2')

# Remove axis labels
axes[0].axis('off')
axes[1].axis('off')

# Display the plots
plt.tight_layout()
plt.show()
```

# Adding area

```python
import geopandas as gpd

# Load the first region
region1 = gpd.read_file('C:\\Users\\LENOVO\\Downloads\\IND_adm\\IND_adm1.shp')

# Load the second region
region2 = gpd.read_file('C:\\Users\\LENOVO\\Downloads\\IND_adm\\IND_adm2.shp')

# Perform the intersection and retain all geometries regardless of type
intersected_object = gpd.overlay(region1, region2, how='intersection', keep_geom_type=False)

# Re-project the geometries to a suitable projected CRS
desired_crs = 'EPSG:3857'
intersected_object = intersected_object.to_crs(desired_crs)

# Calculate the area of the re-projected geometries
intersected_object['area'] = intersected_object['geometry'].area

# Export the resulting GeoDataFrame to a CSV file
intersected_object.to_csv('intersected_data.csv', index=False)
```
